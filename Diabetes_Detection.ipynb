```python
X = dfx.values
Y = dfy.values
print(X.shape)
print(Y.shape)
#X = X[:,1:]
#Y = Y[:,1:]
print(X)
print(Y)
```

```python
def dist(x1,x2):
    return np.sqrt(sum((x1-x2)**2))
```

```python
def knn(X,Y,querypoint,k=6):
    vals = []
    m = X.shape[0]
    for i in range(m):
        d = dist(X[i],querypoint)
        vals.append((d,Y[i]))
    vals = sorted(vals)
    vals = vals[:k]
    vals = np.array(vals)
    vals2=vals[:,1]
    #print(vals2)
    d=sum(vals2)
    if d>4:
        return 1
    else :
        return 0
```

```python
tests = df.values
print (tests.shape)
```

```python
k=tests.shape[0]
#print(k)
k=int(k)

ans=np.full((192,1),0)
for i in range(k):
    a=knn(X,Y,tests[i],6)
    ans[i][0]=a

print(ans)    
ans=pd.DataFrame(ans,columns=['Outcome'])
ans.head()
ans=ans.to_csv('submission.csv',index=False)
```
